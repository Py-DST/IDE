В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def work_experience(arg):
    month_words = ['месяц','месяца','месяцев']
    year_words = ['год','года','лет']
    month = 0
    year = 0
    args_s = []
    if str(arg) == 'nan': return np.nan
    if arg == 'Не указано': return np.nan
    else:
        args_s = arg.split(' ')
        for i in range(7):
            if args_s[i] in month_words:
                month = args_s[i-1]
            if args_s[i] in year_words:
                year = args_s[i-1]
    return int(year)*12 + int(month)

project_1['Опыт работы (месяц)'] = project_1['Опыт работы'].apply(work_experience)
project_1['Опыт работы (месяц)'].median() 

project_1 = project_1.drop(['Опыт работы'], axis=1)    
```

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
def get_town(cities):
    '''
    Функция выделяет из строки название города
    возращаемое значение: str, название горолда либо тип города   
    '''
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 
                      'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    cities_list = cities.split(',')
    cities_list = [x.strip(' ') for x in cities_list]
    cities_type = cities_list[0]
    
    if cities_type in ['Москва','Санкт-Петербург']:
        return cities_type
    elif cities_type in million_cities:
        return 'Город-миллионник'
    else:
        return 'Другие'

project_1['Город'] = project_1['Город, переезд, командировки'].apply(get_town)

#3.4С колько процентов соискателей живут в Санкт-Петербурге?
(project_1['Город'].value_counts()/project_1['Город'].count()*100).round() 

def ready_to_move(move):
    '''
    Функция определяет готовность/не готовность к переезду
    возвращаемое значение: True или False
    '''
    no_move = ['не готов к переезду', 'не готова к переезду']
    move_list = move.split(',')
    move_list = [x.strip(' ') for x in move_list]
    city_type = move_list[1]
    city_type2 = move_list[2]
    if city_type in no_move:
        return False
    if city_type2 in no_move:
        return False
    return True
project_1['Готовность к переезду'] = project_1['Город, переезд, командировки'].apply(ready_to_move)
#(project_1['Готовность к переезду'].value_counts()/project_1['Готовность к переезду'].count()*100).round()

def business_trips(trip):
    '''
    Функция определяет готовность/не готовность к командировкам
    возвращаемое значение: True или False
    '''
    no_trip = ['готов к командировкам','готов к редким командировкам','готова к командировкам','готова к редким командировкам']
    trip_list = trip.split(',')
    trip_list = [x.strip(' ') for x in trip_list]
    for i in range(len(trip_list)):
        if trip_list[i-1] in no_trip:
            return True
        return False
project_1['Готовность к командировкам'] = project_1['Город, переезд, командировки'].apply(business_trips)

#3.4 Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам? Ответ округлите до целого
((project_1.groupby(['Готовность к переезду'])['Готовность к командировкам'].value_counts()/project_1['Готовность к переезду'].count())*100).round()

project_1 = project_1.drop(['Город, переезд, командировки'], axis=1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость

* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
employment_df = ['полная занятость','частичная занятость','проектная работа','стажировка','волонтерство']
schedule_df = ['сменный график','гибкий график','удаленная работа','полный день','вахтовый метод']
for Занятость, График in zip(employment_df, schedule_df):
    project_1[Занятость] = project_1['Занятость'].apply(lambda x: Занятость in x)
    project_1[График] = project_1['График'].apply(lambda x: График in x)

print(project_1[project_1['проектная работа'] & project_1['волонтерство']].shape[0])
print(project_1[project_1['вахтовый метод'] & project_1['гибкий график']].shape[0])

project_1 = project_1.drop('Занятость', axis=1)
project_1 = project_1.drop('График', axis=1)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
#currency_rate = pd.read_csv('/Users/aop/Library/Mobile Documents/com~apple~CloudDocs/IDE/data/Project-1/ExchangeRates.csv', sep=',')
currency_rate = pd.read_csv('data/ExchangeRates.csv', sep=',')
project_1['Обновление резюме'] = pd.to_datetime(project_1['Обновление резюме'], dayfirst=True).dt.date
currency_rate['date'] = pd.to_datetime(currency_rate['date'], dayfirst=True).dt.date

project_1['сумма ЗП'] = project_1['ЗП'].apply(lambda x: x.split(' ')[0])
project_1.loc[:,'сумма ЗП'] = project_1['сумма ЗП'].astype('int')

def get_salary_сurrency(currency):
    """
    функция преобразует наименование валюты к международному сокращенному виду
    возвращаемое значение: str из словаря  
    """
    currency_dict = {
        'USD': 'USD', 'KZT': 'KZT',
        'грн': 'UAH', 'белруб': 'BYN',
        'EUR': 'EUR', 'KGS': 'KGS',
        'сум': 'UZS', 'AZN': 'AZN'
    }
    curr = currency.split(' ')[1].replace('.', '')
    if curr == 'руб':
        return 'RUB'
    else:
        return currency_dict[curr]

project_1['валюта'] = project_1['ЗП'].apply(get_salary_сurrency)

merged = project_1.merge(
    currency_rate,
    left_on=['валюта', 'Обновление резюме'],
    right_on=['currency', 'date'],
    how='left'
)

merged['close'] = merged['close'].fillna(1)
merged['proportion'] = merged['proportion'].fillna(1)
# Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию
# обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)"
merged.loc[:,'сумма ЗП'] = merged['сумма ЗП'].astype('int')
#merged['ЗП (руб)'] = merged['сумма ЗП']*merged['close']/merged['proportion']

project_1['ЗП (руб)'] = merged['сумма ЗП']*merged['close']/merged['proportion']
# Чему равна желаемая медианная заработная плата соискателей в нашей таблице (в рублях)?
round(project_1['ЗП (руб)'].median()/1000)
#project_1.head()
```
